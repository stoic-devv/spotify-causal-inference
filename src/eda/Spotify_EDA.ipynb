```python
skip = []

skip.append((log_mini['skip_1']==True).sum())
skip.append((log_mini['skip_2']==True).sum()-(log_mini['skip_1']==True).sum())
skip.append((log_mini['skip_3']==True).sum()-(log_mini['skip_2']==True).sum())

skip
```

```python
label = ["skip_1", "skip_2", "skip_3"]

plt.barh(label, skip, align='center')
plt.legend()

plt.ylabel('Stages of Skips')
plt.xlabel('Number')
plt.title('Skip Statistics')

plt.show()
```

```python
context_skip = []

context_skip.append(log_mini[log_mini['context_switch'] == 0]['not_skipped'].value_counts()[False])
context_skip.append(log_mini[log_mini['context_switch'] == 0]['not_skipped'].value_counts()[True])
context_skip.append(log_mini[log_mini['context_switch'] == 1]['not_skipped'].value_counts()[False])
context_skip.append(log_mini[log_mini['context_switch'] == 1]['not_skipped'].value_counts()[True])

context_skip
```

```python
context_label = ["Skipped the song in the same playlist","Did not skip the song and in the same playlist",
                 "Skipped the song to a different playlist","Did not skip the song but moved to a different playlist"]

plt.barh(context_label, context_skip, align='center')
plt.legend()

plt.ylabel('Skips VS Playlist')
plt.xlabel('Number')
plt.title('Context Skip Statistics')

plt.show()
```

```python
merge = log_mini.merge(tf_mini, left_on='track_id_clean', right_on='track_id', how='left')
```

```python
corr_mat = log_mini.corr(method='spearman')

hm_dims = (20, 10)

fig, ax = plt.subplots(figsize=hm_dims)
sns.heatmap(corr_mat, square=False, annot=True, linewidths=.4, ax=ax)
ax.set_ylim(len(corr_mat)-0.5, -0.5)
plt.show()
```

```python
corr_mat = merge.corr(method='spearman')

hm_dims = (30, 22)

fig, ax = plt.subplots(figsize=hm_dims)
sns.heatmap(corr_mat, square=False, annot=True, linewidths=.4, ax=ax)
ax.set_ylim(len(corr_mat)-0.5, -0.5)
plt.show()
```
